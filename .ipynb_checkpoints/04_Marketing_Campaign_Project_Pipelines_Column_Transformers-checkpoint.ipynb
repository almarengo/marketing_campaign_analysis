Context
Problem Statement

Customer Personality Analysis is a detailed analysis of a company’s ideal customers. It helps a business to better understand its customers and makes it easier for them to modify products according to the specific needs, behaviors and concerns of different types of customers.

Customer personality analysis helps a business to modify its product based on its target customers from different types of customer segments. For example, instead of spending money to market a new product to every customer in the company’s database, a company can analyze which customer segment is most likely to buy the product and then market the product only on that particular segment.

Content
Attributes

People

- ID: Customer's unique identifier
- Year_Birth: Customer's birth year
- Education: Customer's education level
- Marital_Status: Customer's marital status
- Income: Customer's yearly household income
- Kidhome: Number of children in customer's household
- Teenhome: Number of teenagers in customer's household
- Dt_Customer: Date of customer's enrollment with the company
- Recency: Number of days since customer's last purchase
- Complain: 1 if the customer complained in the last 2 years, 0 otherwise

Products

- MntWines: Amount spent on wine in last 2 years
- MntFruits: Amount spent on fruits in last 2 years
- MntMeatProducts: Amount spent on meat in last 2 years
- MntFishProducts: Amount spent on fish in last 2 years
- MntSweetProducts: Amount spent on sweets in last 2 years
- MntGoldProds: Amount spent on gold in last 2 years

Promotion

- NumDealsPurchases: Number of purchases made with a discount
- AcceptedCmp1: 1 if customer accepted the offer in the 1st campaign, 0 otherwise
- AcceptedCmp2: 1 if customer accepted the offer in the 2nd campaign, 0 otherwise
- AcceptedCmp3: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise
- AcceptedCmp4: 1 if customer accepted the offer in the 4th campaign, 0 otherwise
- AcceptedCmp5: 1 if customer accepted the offer in the 5th campaign, 0 otherwise
- Response: 1 if customer accepted the offer in the last campaign, 0 otherwise

Place

- NumWebPurchases: Number of purchases made through the company’s website
- NumCatalogPurchases: Number of purchases made using a catalogue
- NumStorePurchases: Number of purchases made directly in stores
- NumWebVisitsMonth: Number of visits to company’s website in the last month

## Categorical columns

Let's bucket some of the categories together:
- 'Second Cycle' can be incorporated into 'Master' for Education
- 'Together' can be consider as 'Married' in Marital Status
- 'Alone' can be consider as 'Single' in Marital Status

We can also drop 'YOLO' and 'Absurd' that are only few

```python
df.drop(index=df[df['Marital_Status'] == 'YOLO'].index, inplace=True)
df.drop(index=df[df['Marital_Status'] == 'Absurd'].index, inplace=True)
```

```python
df['Education'] = np.where(df['Education'] == '2n Cycle', 'Master', df['Education'])
df['Marital_Status'] = np.where(df['Marital_Status'] == 'Together', 'Married', df['Marital_Status'])
df['Marital_Status'] = np.where(df['Marital_Status'] == 'Alone', 'Single', df['Marital_Status'])
```

Let's save the categorical columns into a variable, we'll use them later in the pipeline

```python
categorical_columns = df.select_dtypes(include="object_").columns.tolist()
```

Let's do some simple feature engineering on the datetime column and extract month and year.

```python
df['Dt_MONTH'] = df['Dt_Customer'].dt.month
df['Dt_YEAR'] = df['Dt_Customer'].dt.year
to_remove.append('Dt_Customer')
df.drop(columns=to_remove, axis=1, inplace=True)
```

We had alread a variable for the numerical column but we should add the newly created 2 columns.

```python
num_columns.extend(['Dt_MONTH', 'Dt_YEAR'])
```

```python
df.head()
```

## Column Transformer and Pipeline

We can now create the X (features) and y (target)

```python
X = df.drop(columns='Response')
y = df['Response']
```

```python
print(X.shape, y.shape)
```

Below are all the numerical columns.....

```python
num_columns.remove('Response')
num_columns
```

.... and these are the categorical ones.

```python
categorical_columns
```

```python
X
```

We create 2 pipelines, one for the numeric columns and one for the categorical columns.
- On the numeric column we will apply a Standard Scaler
- On the categorical column we will apply the One Hot Encoder

NOTE: On the numeric columns pipeline it would be nice to a step that calls a class to remove outlier (we did this in notebook 3)

We then apply the column trasformer assigning each pipeline to the corresponding columns. All together, this will be a pre-processing pipeline.

```python
numeric_transformer = Pipeline(steps=[("scaler", StandardScaler())])
categorical_transformer = OneHotEncoder(handle_unknown="ignore")

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_columns),
        ("cat", categorical_transformer, categorical_columns)])
```

We then create another pipeline with 2 steps: the pre-processing and the model. We will fit a logistic regression.

```python
# Pipleine with model
clf = Pipeline(steps=[("preprocessor", preprocessor), ("classifier", LogisticRegression(max_iter=500))])
```

Lastly, before fitting the whole pipeline, we have to split the dataset into train and test.

```python
# Split the data with 33% to test and keepinf the same balance that we have in y (Target)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, stratify=y)
```

```python
print(X_train.shape, y_train.shape)
```

```python
print(X_test.shape, y_test.shape)
```

## Fit Logistic Regression Model

```python
clf.fit(X_train, y_train)
print("model score: %.3f" % clf.score(X_test, y_test))
```

The accuracy on the test set is not bad.

Below is a nice html visual of the whole pipeline.

```python
set_config(display="diagram")
clf
```

## Confusion Matrix

```python
y_test_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, y_test_pred))
```

Again, we see that the precision of the model is not bad but the recall is low (a lot of False Negatives).

```python
confusion_df = pd.DataFrame(confusion_matrix(y_test, y_test_pred))
confusion_df.index = ['Actual 0', 'Actual 1']
confusion_df.columns = ['Predicted 0', 'Predicted 1']
confusion_df
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(confusion_df, annot=True, fmt='d', linewidths=0.5) 
plt.yticks(rotation=0)
plt.xticks(rotation=45)
plt.show()
```

## GridSearchCV

As we did in Notebook 3 for pre-pruning, we can also apply GridSearchCV on the pipeline.

Below we search between 2 different data normalizer and different C values (inverse of the regularization strength) to find the best combination.

```python
param_grid = {
    "preprocessor__num__scaler": [MinMaxScaler(), StandardScaler()],
    "classifier__C": [0.1, 1.0, 10, 100],
}

grid_search = GridSearchCV(clf, param_grid, cv=10)
grid_search
```

```python
grid_search.fit(X_train, y_train)

print("Best params:")
print(grid_search.best_params_)
```

It looks like the best pipeline uses the StandardScaler and a Logistic Regression with C = 0.1 on cross validation of 10.

```python
print(f"Train CV score: {grid_search.best_score_:.3f}")
```

```python
print(
    (
        "best logistic regression from grid search: %.3f"
        % grid_search.score(X_test, y_test)
    )
)
```

On the test set the accuracy is very similar to what we had before.

Below is the table that shows the top 5 best combinations, all very close together. 

```python
cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results = cv_results.sort_values("mean_test_score", ascending=False)
cv_results[
    [
        "mean_test_score",
        "std_test_score",
        "param_preprocessor__num__scaler",
        "param_classifier__C",
    ]
].head(5)
```

```python
y_test_pred = grid_search.predict(X_test)
```

```python
print(classification_report(y_test, y_test_pred))
```
